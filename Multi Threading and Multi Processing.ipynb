```python
# with threading
import threading
import time

def task():
    print("Task started")
    time.sleep(3)
    print("Task finished")

t = threading.Thread(target=task)
t.start()

print("I can run while task runs!")
```

```python
# with out thread
import threading
import time

def print_numbers():
    for i in range(5):
        
        print(i)
        time.sleep(2)
        
def print_variabls():
    for i in "abcde":
        print(i)
        time.sleep(2)

t=time.time()

print_numbers()
print_variabls()

print(time.time()-t)
```

```python
# with  thread
import threading
import time

def print_numbers():
    for i in range(5):
        
        print(i)
        time.sleep(2)
        
def print_variabls():
    for i in "abcde":
        print(i)
        time.sleep(2)

t=time.time()
t1=threading.Thread(target=print_numbers)
t2=threading.Thread(target=print_variabls)

t1.start()
t2.start()

t1.join()
t2.join()


print(time.time()-t)
```
