```python
import numpy as np

arr1=np.array([1,2,3,4,5])
print(arr1)
print(type(arr1))
print(arr1.shape)

arr1.reshape(1,5)
# 1 row 5 cols
print(arr1)
```

```python
arr=np.arange(0,10,2).reshape(5,1)

print(arr)
```

```python
np.ones((3,4))
```

```python
arr1=np.array([10,20,30,40,50])
arr2=np.array([50,60,70,80,90])

print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
```

```python
arr1=np.arange(0,10)
print(arr1)
print(np.sqrt(arr1))
print(np.exp(arr1))
print(np.sin(arr1))
print(np.log(arr1))
```

```python
# slicing and indexing

arr1=np.array([[1,2,3,4],[5,6,7,8],[7,8,9,10]])
print(arr1)
```

```python
arr1[0][0]
arr1[1:,2:]
```

```python
arr1[0][0]=100
print(arr1)
```

```python
arr1[1:]=100
print(arr1)
```
