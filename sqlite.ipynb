```python
import sqlite3
 
connection=sqlite3.connect("example.db")
```

```python
cursor=connection.cursor()
```

```python
cursor.execute('''
               create table employee(id primarykey int(20),
               name varchar(20),
               age int(20)
               )
               ''')

connection.commit()
```

```python
cursor.execute('''
               insert into employee(id,name,age)
               values(1,"sandy",10)
               ''')
connection.commit()
```

```python
cursor.execute('''
               select * from employee
               ''')

rows=cursor.fetchall()

for row in rows:
    print(row)
```
